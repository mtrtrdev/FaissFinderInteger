```python
# 必要なライブラリのインポート
import numpy as np  # 数値計算ライブラリのNumPyをインポート
import faiss        # Faissライブラリをインポート

# データの準備
d = 128  # ベクトルの次元数（例：128次元ベクトル）
nb = 10000  # ベクトルの数（例：10000個のベクトル）
np.random.seed(1234)  # 再現性のためのシード設定（乱数の初期化）
xb = np.random.random((nb, d)).astype('float32')  # ランダムなベクトルを生成し、浮動小数点に変換

# 最初の10件のベクトルデータを参照する
first_10_vectors = xb[:10]
print(first_10_vectors)

# インデックスの作成と学習
index = faiss.IndexFlatL2(d)  # L2距離（ユークリッド距離）を使用するフラットインデックスを作成
index.add(xb)  # 生成したベクトルをインデックスに追加
print("インデックスに格納されているベクトルの数:", index.ntotal)  # インデックスに格納されているベクトルの数を表示

# ベクトル検索の実行
nq = 1  # クエリ（検索するベクトル）の数
xq = np.random.random((nq, d)).astype('float32')  # 検索用のクエリベクトルを生成

k = 4  # 上位k個の最近傍ベクトルを検索（例：4個）
D, I = index.search(xq, k)  # 検索を実行し、距離（D）とインデックス（I）を取得

print("検索クエリ:", xq)
print("最近傍ベクトルのインデックス:", I)  # 最近傍ベクトルのインデックスを表示
print("最近傍までの距離:", D)  # 最近傍ベクトルまでの距離を表示
```
